```python
import sys
import pandas as pd
import numpy as np
from PyQt5.QtWidgets import QFileDialog
from PyQt5 import Qt
from window import Ui_MainWindow
from PandasModel import PandasModel
from PyQt5 import QtWidgets, uic
from modul_predict import predict_All_Fragments_Variety
from modul_predict_al_cr import predict_Al_Cr_fragment_and_sort
 
class MsgBox(Qt.QDialog):
    def __init__(self):
        super().__init__()
        layout = Qt.QVBoxLayout(self)
        layout.setObjectName("Dialog")
        self.label = Qt.QLabel("Ошибка классификации. \n Формат данных не верен.")
        layout.addWidget(self.label)
        close_btn = Qt.QPushButton("Ok")
        close_btn.clicked.connect(self.close)
        layout.addWidget(close_btn)


class TestApp(Ui_MainWindow):
    def __init__(self, dialog):
        Ui_MainWindow.__init__(self)
        self.setupUi(dialog)    
        self.header_data_all  = ['Al','Ba','Cr','Cu','Fe','Li','Mn','Na','Ni','Pb','Rb','Sr','Ti','Zn','Ca','K_','Mg']
        self.header_data_al_cr  = ['Al', 'Cr']
        self.df = pd.DataFrame(None, columns=None)     
        self.df_predict = pd.DataFrame(None, columns=None)
        
        self.pushButton.clicked.connect(self.load_File)
        self.pushButton_2.clicked.connect(self.file_save)
        self.pushButton_3.clicked.connect(self.clear)
        self.pushButton_4.clicked.connect(self.df_init_all)
        self.pushButton_6.clicked.connect(self.df_init_al_cr)
        self.pushButton_5.clicked.connect(self.data_predict)
 
        self.msg = MsgBox()
        self.msg.setFixedSize(200, 100)    
    
    def clear(self):
        self.df = pd.DataFrame(None, columns=None)
        self.df_predict = pd.DataFrame(None, columns=None)
        self.showDf(self.df)

        
    def showDf(self, df):
        model = PandasModel(df)
        self.tableView.setModel(model)   
    
    
    def load_File(self):      
        fileName, _ = QtWidgets.QFileDialog.getOpenFileName(None, 'Dialog Title', "", "CSV Files (*.csv)")      
        df = pd.read_csv(fileName, sep=";", header=None)
        if( len(df.columns) == 17): 
            self.df = pd.DataFrame(df.values, columns = self.header_data_all )
        elif( len(df.columns)== 2 ):  
            self.df = pd.DataFrame(df.values, columns = self.header_data_al_cr )
        
        else: self.df =  pd.DataFrame(df.values, columns = None)
        self.showDf(self.df)
        
        
    def data_predict(self):
        if(not(self.df.isnull().any().any())):
            try:
                if( len(self.df.columns)==17): 
                    self.df_predict = predict_All_Fragments_Variety(self.df)
                    self.showDf(self.df_predict)
                elif( len(self.df.columns)==2 ): 
                    self.df_predict = predict_Al_Cr_fragment_and_sort(self.df)    
                    self.showDf(self.df_predict)
                else: self.msg.show()
            except: self.msg.show()
        else:
            self.msg.show()
        
   
    def df_init_all(self):
        self.df = pd.DataFrame({
            'Al': [self.lineEdit.text()],
            'Ba': [self.lineEdit_2.text()],
            'Cr': [self.lineEdit_3.text()],
            'Cu': [self.lineEdit_4.text()],
            'Fe': [self.lineEdit_5.text()],
            'Li': [self.lineEdit_6.text()],
            'Mn': [self.lineEdit_7.text()],
            'Na': [self.lineEdit_8.text()],
            'Ni': [self.lineEdit_9.text()],
            'Pb': [self.lineEdit_10.text()],
            'Rb': [self.lineEdit_11.text()],
            'Sr': [self.lineEdit_12.text()],
            'Ti': [self.lineEdit_13.text()],
            'Zn': [self.lineEdit_14.text()],
            'Ca': [self.lineEdit_15.text()],
            'K_': [self.lineEdit_16.text()],
            'Mg': [self.lineEdit_17.text()],
                  })
        self.showDf(self.df)
        
    def df_init_al_cr(self):
        self.df = pd.DataFrame({
            'Al': [self.lineEdit_20.text()],
            'Cr': [self.lineEdit_21.text()],
                  })
        self.showDf(self.df)

    def file_save(self):
        if(not(self.df.empty)):
            points_file, _ = QFileDialog.getSaveFileName(caption="Save Points", filter='CSV Files (*.csv)')
            if len(points_file) != 0:
                self.df_predict.to_csv(points_file, sep=";", index=None)

def main():
    app = QtWidgets.QApplication(sys.argv)
    dialog = QtWidgets.QMainWindow()
    test_1 = TestApp(dialog)
    dialog.show()
    sys.exit(app.exec_())
    
main()
```
